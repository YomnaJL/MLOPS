#### 6. FEATURE ENGINEERING - CRÃ‰ATION DE NOUVELLES VARIABLES

```python
#2.3 Analyse temporelle COMPLÃˆTE 

# Conversion date
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format="%m/%d/%Y %I:%M:%S %p")

# Features temporelles
df['Year']      = df['DATE OCC'].dt.year
df['Month']     = df['DATE OCC'].dt.month
df['Day']       = df['DATE OCC'].dt.day
df['Hour']      = df['TIME OCC'] // 100
df['Weekday']   = df['DATE OCC'].dt.weekday  # 0 = lundi, 6 = dimanche
df['is_weekend']= df['Weekday'].isin([5, 6])

# Noms franÃ§ais des jours (pour les graphs)
day_names_fr = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
df['DayName'] = df['Weekday'].map(dict(enumerate(day_names_fr)))

# Noms des mois en franÃ§ais
month_names_fr = ['Jan', 'FÃ©v', 'Mar', 'Avr', 'Mai', 'Juin',
                  'Juil', 'AoÃ»', 'Sep', 'Oct', 'Nov', 'DÃ©c']
df['MonthName'] = df['Month'].map(dict(enumerate(month_names_fr, 1)))

# =============================================================================
# GRAPHIQUES
# =============================================================================

fig = plt.figure(figsize=(20, 12))

# 1. Par annÃ©e
plt.subplot(2, 3, 1)
sns.countplot(data=df, x='Year', palette="dark", order=sorted(df['Year'].unique()))
plt.title("Ã‰volution annuelle des crimes dÃ©clarÃ©s", fontsize=14, fontweight='bold')
plt.xlabel("AnnÃ©e")
plt.ylabel("Nombre de crimes")

# 2. Par mois (saisonnalitÃ©)
plt.subplot(2, 3, 2)
monthly = df['MonthName'].value_counts().reindex(month_names_fr)
monthly.plot(kind='bar', color='orange', ax=plt.gca())
plt.title("RÃ©partition par mois (toutes annÃ©es confondues)", fontsize=14, fontweight='bold')
plt.xlabel("Mois")
plt.ylabel("Nombre de crimes")
plt.xticks(rotation=45)

# 3. Par jour de la semaine â† NOUVEAU !
plt.subplot(2, 3, 3)
weekday_order = day_names_fr
sns.countplot(data=df, x='DayName', order=weekday_order, palette="viridis")
plt.title("RÃ©partition par jour de la semaine", fontsize=14, fontweight='bold')
plt.xlabel("Jour")
plt.ylabel("Nombre de crimes")
plt.xticks(rotation=45)

# 4. Par heure (24h)
plt.subplot(2, 3, 4)
sns.countplot(data=df, x='Hour', palette="coolwarm")
plt.title("RÃ©partition par heure de la journÃ©e", fontsize=14, fontweight='bold')
plt.xlabel("Heure (format 24h)")
plt.ylabel("Nombre de crimes")

# 5. Week-end vs Semaine
plt.subplot(2, 3, 5)
weekend_counts = df['is_weekend'].map({True: 'Week-end', False: 'Semaine'}).value_counts()
weekend_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], ax=plt.gca())
plt.title("Proportion week-end vs semaine", fontsize=14, fontweight='bold')
plt.ylabel("")

# 6. Heatmap jour Ã— heure (le plus beau !)
plt.subplot(2, 3, 6)
heatmap_data = df.groupby(['DayName', 'Hour']).size().unstack(fill_value=0)
heatmap_data = heatmap_data.reindex(day_names_fr)  # ordre lundi â†’ dimanche
sns.heatmap(heatmap_data, cmap="YlOrRd", linewidths=.5, cbar_kws={'label': 'Nombre de crimes'})
plt.title("Heatmap : Crimes par jour et heure", fontsize=14, fontweight='bold')
plt.xlabel("Heure")
plt.ylabel("Jour de la semaine")

plt.tight_layout()
plt.show()
```

```python
# VISUALISATIONS SIMPLES & EFFICACES 


plt.figure(figsize=(20, 12))

# 1. Top 10 lieux oÃ¹ les crimes se passent
plt.subplot(2, 3, 1)
top10_lieux = df['Premis Desc'].value_counts().head(10)
top10_lieux.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title("Top 10 lieux de crimes", fontsize=14, fontweight='bold')
plt.xlabel("Nombre de crimes")
plt.gca().invert_yaxis()  # le plus frÃ©quent en haut

# 2. Top 10 quartiers (divisions LAPD)
plt.subplot(2, 3, 2)
top10_area = df['AREA NAME'].value_counts().head(10)
top10_area.plot(kind='barh', color='salmon', edgecolor='black')
plt.title("Top 10 quartiers les plus criminogÃ¨nes", fontsize=14, fontweight='bold')
plt.xlabel("Nombre de crimes")
plt.gca().invert_yaxis()

# 3. Carte simple (tous les points)
plt.subplot(2, 3, 3)
df_plot = df[(df['LAT'] != 0) & (df['LON'] != 0)]
plt.scatter(df_plot['LON'], df_plot['LAT'], s=0.5, alpha=0.5, c='red')
plt.title("Tous les crimes sur la carte de Los Angeles", fontsize=14, fontweight='bold')
plt.axis('off')

# 4. Pie chart â€“ Rue vs IntÃ©rieur vs Autre
plt.subplot(2, 3, 4)
lieu_simple = df['Premis Desc'].str.contains('STREET|SIDEWALK|ALLEY|PARKING|GARAGE', case=False, na=False)
counts = lieu_simple.value_counts()
labels = ['Rue / Trottoir / Parking', 'Lieu fermÃ© (maison, magasin, etc.)']
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['orange', 'lightblue'], startangle=90)
plt.title("Crimes en extÃ©rieur vs intÃ©rieur")

# 5. Top 5 lieux + Top 5 quartiers (combinÃ©)
plt.subplot(2, 3, 5)
top5_lieux = df['Premis Desc'].value_counts().head(5)
top5_area = df['AREA NAME'].value_counts().head(5)
combined = pd.concat([top5_lieux, top5_area])
combined.plot(kind='bar', color=['C0' if i<5 else 'C1' for i in range(10)])
plt.title("Top 5 lieux + Top 5 quartiers")
plt.xticks(rotation=45)
plt.ylabel("Nombre de crimes")

# 6. Bonus : densitÃ© trÃ¨s simple (heatmap rapide)
plt.subplot(2, 3, 6)
plt.hist2d(df_plot['LON'], df_plot['LAT'], bins=100, cmap='Reds', alpha=0.8)
plt.colorbar(label='DensitÃ©')
plt.title("Carte de chaleur simple des crimes")
plt.axis('off')

plt.suptitle("ðŸ”¥ VISUALISATIONS SIMPLES : LIEUX & LOCALISATION ðŸ”¥", 
             fontsize=18, fontweight='bold', y=1)
plt.tight_layout()
plt.show()
```
